# 2. What do you think about the students' parental level of education?

```python
sns.countplot(x='parental level of education',data=data)
plt.xticks(rotation=90)
plt.title("Students' parental level of education")
```

Most of the students have college as their parental level of education

```python
sns.countplot(x='race/ethnicity',hue='parental level of education',data=data)
plt.legend(loc=(1.02,0))
plt.title("Students' parental level of education")
```

Considering each race, majority of group C and E students have associate's degree as their parental level of education. But for group B and A high school is the parental level of education that most of students have. whereas for group D it is college.

```python
sns.countplot(x='gender',hue='parental level of education',data=data)
plt.xticks(rotation=90)
plt.legend(loc=(1.02,0))
plt.title("Students' parental level of education")
```

From the graph it is clear that for male students as well as female students the top bar of parental level of education is shown by college.
